# Fix Data #

```python
# Check before converting float columns to int

def check_float_for_int(df, column):
    mask = df[column].notna()
    are_equal = np.array_equal(df[column][mask], df[column].astype('Int64')[mask])

    if are_equal:
        print(f"All non-missing values in the {column} column are integers: proceed to convert")
    else: 
        print(f"Some non-missing values in the {column} column are not integers: DO NOT convert")
    
check_float_for_int(df_cars, 'model_year')
check_float_for_int(df_cars, 'cylinders')
check_float_for_int(df_cars, 'odometer')
check_float_for_int(df_cars, 'is_4wd')
```

```python
# Convert float columns to int

df_cars['model_year'] = df_cars['model_year'].astype('Int64')
df_cars['cylinders'] = df_cars['cylinders'].astype('Int64')
df_cars['odometer'] = df_cars['odometer'].astype('Int64')
df_cars['is_4wd'] = df_cars['is_4wd'].astype('Int64')

display(df_cars.info())
df_cars
```

```python
# Check if strings are all lowercase

def check_lowercase(df, column):
    if df[column].str.islower().all():
        print(f"All non-missing values in the {column} column are lowercase")
    else: 
        print(f"Some non-missing values in the {column} column are NOT lowercase")
    
check_lowercase(df_cars, 'model')
check_lowercase(df_cars, 'condition')
check_lowercase(df_cars, 'fuel')
check_lowercase(df_cars, 'transmission')
check_lowercase(df_cars, 'type')
check_lowercase(df_cars, 'paint_color')
```

```python
# Convert type column to lowercase

df_cars['type'] = df_cars['type'].str.lower()

check_lowercase(df_cars, 'type')
```

```python
# Convert missing values to 0s for the is_4wd column

df_cars['is_4wd'] = df_cars['is_4wd'].fillna(0)

# Check to see if new values make sense

display(df_cars['is_4wd'].value_counts())
print()
display(df_cars.groupby('model')['is_4wd'].mean().sort_values(ascending=False).head(10))
print()
df_cars.groupby('model')['is_4wd'].mean().sort_values(ascending=False).tail(10)
```

Whereas vehicles with very high values in the is_4wd column are mostly trucks and SUVs, those with very low values tend to be compact and sports cars, so the new values appear to make sense.

```python
# Extract make and model/design from the model column

df_cars['make'] = df_cars['model'].str.split().str[0]
df_cars['design'] = df_cars['model'].str.split().str[1:].str.join(' ')

df_cars
```

# Expore and Plot Data #

```python
# Create a function to plot histograms

def histogram(df, column, label, bins=25, angle=0, cutoff=0, cutoff_value=0):
    display(df[column].describe())
    
    if cutoff:
        if cutoff == '<':
            plt.hist(df[df[column] <= cutoff_value][column], bins=bins)
        elif cutoff == '>':
            plt.hist(df[df[column] >= cutoff_value][column], bins=bins)
        else:
            print("Enter a cutoff direction")
    else:
        plt.hist(df[column], bins=bins)
    
    plt.xlabel(label)
    plt.ylabel('Number of Observations')
    plt.title(label)
    plt.xticks(rotation=angle)
    plt.show()
```

```python
# Examine the distribution of price

histogram(df_cars, 'price', 'Price', cutoff='<', cutoff_value=50000)
```

```python
# Examine the distribution of odometer

histogram(df_cars, 'odometer', 'Odometer Reading (Miles)', angle=30, cutoff='<', cutoff_value=300000)
```

# Investigate Relationships Among Variables #

```python
# Create a function to plot bar charts of variable averages

def bar_chart_avg(df, x_column, y_column, x_label, y_label, angle=0):
    df_clean = df[[x_column, y_column]].dropna()
    df_grouped = df_clean.groupby([x_column])[y_column].agg(['mean']).sort_values(by='mean', ascending=False).reset_index()
    
    plt.bar(df_grouped[x_column], df_grouped['mean'])
    plt.xlabel(x_label)
    plt.ylabel(f'Average {y_label}')
    plt.title(f'Average {y_label} by {x_label}')
    plt.xticks(rotation=angle)
    plt.show()
```

```python
# Plot price, days listed, and odometer by make

bar_chart_avg(df_cars, 'make', 'price', 'Manufacturer', 'Price', 90)
bar_chart_avg(df_cars, 'make', 'days_listed', 'Manufacturer', 'Days Listed', 90)
bar_chart_avg(df_cars, 'make', 'odometer', 'Manufacturer', 'Odometer', 90)
```

Mercedez-Benz sells for much higher average prices than the rest followed by Ram, Cadillac, and GMC with Hyundai and Acura selling for least. BMW, Cadillac, and Kia take the longest to sell, though not by much. Acuras have noticeably more miles on their odometers than the rest followed by Toyota, GMC, and Honda with Kia and Nissan at the bottom.

```python
# Plot price, days listed, and odometer by condition

bar_chart_avg(df_cars, 'condition', 'price', 'Condition', 'Price', 90)
bar_chart_avg(df_cars, 'condition', 'days_listed', 'Condition', 'Days Listed', 90)
bar_chart_avg(df_cars, 'condition', 'odometer', 'Condition', 'Odometer', 90)
```

Condition has a very clear effect on prices. New vehicles sell for noticeably more than the rest followed by like new, excellent, and good with salvage and fair noticeably lower. New cars sellslightly faster than the rest, but not by much. Predictably, fair vehicles have noticeably higher odometer readings followed by salvage, good, and excellent with like new and new at the bottom. 

```python
# Plot price, days listed, and odometer by vehicle type

bar_chart_avg(df_cars, 'type', 'price', 'Vehicle Type', 'Price', 90)
bar_chart_avg(df_cars, 'type', 'days_listed', 'Vehicle Type', 'Days Listed', 90)
bar_chart_avg(df_cars, 'type', 'odometer', 'Vehicle Type', 'Odometer', 90)
```

Buses have the highest prices, followed by trucks and pickups with sedans and hatchbacks selling for the least. Buses also take noticeably longer to sell than other vehicles followed by vans and wagons. Mini-vans have the most odometer miles followed by vans, wagons, and trucks with convertibles and coupes coming in considerably lower than the rest.

Price, odometer, make, type, and model year are all worthwhile variables to investigate.
